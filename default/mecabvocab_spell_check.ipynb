```python
f = open("final_vocabs.txt", 'r')

lines = f.readlines()
print(len(lines))
f.close()

temp = ""

change_words = []
error_words = []

for i, word in enumerate(lines):
    if i % 100 == 0: print("current : {}, number of changed word : {}".format(i, len(change_words)))
    if word[:2] == "##":
        temp = word[2:-1]
    else:
        temp = word[:-1]
    try:
        result = spell_checker.check(temp)
        checked_word = result.as_dict()['checked']
        if temp == checked_word: pass
        else: change_words.append((i, checked_word))
    except Exception as e:
        print(e,'\nerror index : {}, word : {}'.format(i, word))
        error_words.append((i, word))
    
    temp = ""

print(len(change_words))
```
