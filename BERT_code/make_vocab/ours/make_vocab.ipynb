### Mecab Vocab 처리(사용 x)

```python
f = open("rsc/my_conf/FinalMecabVocab.txt", 'r')
lines = f.readlines()
print("Total Vocab size : ", len(lines))
f.close()
```

```python
# 모든 숫자 제거
temp_list = []
def check_digit(text):
    if text[:2] == "##": text = text[2:]
    return text.isdigit()
        
for i in lines:
    if not check_digit(i[:-1]):
        temp_list.append(i)
print("Number Removed Vocab size  : ", len(temp_list))
```

## English Vocab 만들기
영어만 남기고 모두 제거 -> **wordpiece**사용하여 english vocab 만들기

```python
import re
f = open(FINAL_TXT_SENTENCE_PATH, 'r')

full_text = f.read()
# 영어 띄어쓰기만 남기고 모두 제거
full_text = re.sub('[^a-zA-Z ]',' ',full_text)
print("Complete : 영어 띄어쓰기만 남기고 모두 제거")

# 다중 띄어쓰기 제거
full_text = re.sub(' +', ' ', full_text)
print("Complete : 다중 띄어쓰기 제거")

# # 모두 소문자로
# full_text.lower()
```

```python
g = open(FINAL_TXT_ONLY_ENG_PATH, 'w')
g.write(full_text)
g.close()
f.close()
```
